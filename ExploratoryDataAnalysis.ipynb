```python
%matplotlib inline
import matplotlib.pyplot as plt
import os, sys
from PIL import Image

from utils.helpers import *
```

```python
# training images:
root_training_dir = "dataset/training/"
image_dir = root_training_dir + "images/"
gt_dir = root_training_dir + "groundtruth/"

files_train = os.listdir(image_dir)

n_train = len(files_train)
print(
    "Loading all the images of the training dataset: {} images and groundtruths".format(
        n_train
    )
)

imgs = [load_image(image_dir + files_train[i]) for i in range(n_train)]
gt_imgs = [load_image(gt_dir + files_train[i]) for i in range(n_train)]
```

```python
# Test:
root_test_dir = "dataset/test_set_images/"
folder_names = os.listdir(root_test_dir)
n_test = len(folder_names)  # Load maximum 20 images
print("Loading all the images of the test dataset: {} images".format(n_test))
imgs_test = [
    load_image(os.path.join(root_test_dir, folder_names[i], f"{folder_names[i]}.png"))
    for i in range(n_test)
]
```

```python
all_img_train_shapes = [img_train.shape for img_train in imgs]
all_gt_shapes = [gt_img.shape for gt_img in gt_imgs]
all_imgs_test_shapes = [img_test.shape for img_test in imgs_test]
# check that all the images in a same folder have the same dimension
if (
    len(set(all_img_train_shapes)) == 1
    and len(set(all_gt_shapes)) == 1
    and len(set(all_imgs_test_shapes)) == 1
):
    print(f"All training images have the same dimensions: {all_img_train_shapes[0]}")
    print(f"All ground truth images have the same dimensions: {all_gt_shapes[0]}")
    print(f"All test images have the same dimensions: {all_imgs_test_shapes[0]}")
else:
    print("The images do not all have the same dimensions in a same folder!")
```
